```python
import cv2
import numpy as np
import os
import warnings

warnings.filterwarnings("ignore")
```

## 1. Read Image and show each image on each window

```python
image_dir = "./images/jpg"
```

```python
for filename in os.listdir(image_dir):
    img = cv2.imread(os.path.join(image_dir, filename))
    cv2.imshow(filename, img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

## 2. Separate RGB image to 3 layers and show ench layer on ech window

```python
img = cv2.imread("images/jpg/image_1.jpg")

b, g, r = cv2.split(img)

cv2.imshow("Red", r)
cv2.imshow("Green", g)
cv2.imshow("Blue", b)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 3. Convert RGB image to grey and show it

```python
img = cv2.imread("images/jpg/image_1.jpg")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Grayscale", img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 4. Rotate an image in 50 times: In each step, rotated 15 degrees and zoom out to 90% of its size and then show it on the same window

```python
img = cv2.imread("images/jpg/image_1.jpg")
h, w = img.shape[:2]
center = (w // 2, h // 2)

total_frames = 100
angle = 0
angle_step = 15

for i in range(total_frames):
    # Giai đoạn 1: 0 <= i < 50 → thu nhỏ (scale 1.0 -> 0.3)
    # Giai đoạn 2: i >= 50 → phóng to (scale 0.3 -> 1.0)
    if i < 50:
        progress = i / 49.0  # 0 → 1 trong 50 bước
        scale = 1.0 - 0.7 * progress  # từ 1.0 xuống 0.3
    else:
        progress = (i - 50) / 49.0  # 0 → 1 trong 50 bước
        scale = 0.3 + 0.7 * progress  # từ 0.3 lên 1.0

    angle += angle_step
    M = cv2.getRotationMatrix2D(center, angle, scale)
    rotated = cv2.warpAffine(img, M, (w, h))
    cv2.imshow("Rotating", rotated)

    if cv2.waitKey(100) == 27:
        break

cv2.destroyAllWindows()
```

## 5. Read an image, then crop image with 1⁄4 its size from image center and then show it.

```python
img = cv2.imread("images/jpg/image_1.jpg")
h, w = img.shape[:2]
crop_w, crop_h = w // 2, h // 2

x1 = (w - crop_w) // 2
y1 = (h - crop_h) // 2

cropped = img[y1 : y1 + crop_h, x1 : x1 + crop_w]
cv2.imshow("Original", img)
cv2.imshow("Cropped", cropped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
