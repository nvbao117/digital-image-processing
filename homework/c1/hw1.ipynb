```python
import cv2
from PIL import Image
```

```python
import os
```

```python
input_folder = "./original"
output_folder = "./images"
os.makedirs(os.path.join(output_folder, "png"), exist_ok=True)
os.makedirs(os.path.join(output_folder, "jpg"), exist_ok=True)
os.makedirs(os.path.join(output_folder, "bmp"), exist_ok=True)
```

```python
TARGET_SIZE = (800, 600)

for filename in os.listdir(input_folder):
    if filename.lower().endswith((".jpg", ".png", ".jpeg")):
        img = Image.open(os.path.join(input_folder, filename))
        img = img.resize(TARGET_SIZE, Image.LANCZOS)
        name = os.path.splitext(filename)[0]
        img.save(os.path.join(output_folder, f"./png/{name}.png"))
        img.save(os.path.join(output_folder, f"./jpg/{name}.jpg"))
        img.save(os.path.join(output_folder, f"./bmp/{name}.bmp"))
        print(f"Done: {filename}")
```
